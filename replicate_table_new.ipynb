```python
#get equity RIC
eqt_RIC=pd.read_csv('eqt_RIC.csv',usecols=['Type','RIC'])
eqt_RIC.head()
```

```python
sector=pd.read_excel('hkmkt_equity_industry_name (1).xlsm','Sheet1',usecols=['Type','TR1N'])

mktcap = eqt_mktcap.reindex(columns=['Name','2013-12-31','2014-12-31','2015-12-31','2016-12-30','2017-12-29','2018-12-31'])
mktcap['Name']=eqt_RIC['RIC']
mktcap['Economic Sector']=sector['TR1N']
mktcap.columns=['Name','2013','2014','2015','2016','2017','2018','Economic Sector']
mktcap['Name']=mktcap['Name'].astype(str).str.replace('.0','',regex=False)

mktcap
```

```python
def find_matching_firm(mktdf,industry,mktcap,year):
    same = mktdf[mktdf['Economic Sector'] == industry]
    min_diff = float('inf')
    match = 'Name'
    for index, row in same.iterrows():
        diff = abs(row[year]-mktcap)
        if diff < min_diff:
            min_diff = diff
            match = row['Name']
    return match
```

```python
peers=pd.DataFrame(columns=['IPO','IPO return','MFirm','MFirm return'])

for index, row in data_ipo.iterrows():
    ipo = row['Stock']
    a = row['Industry']
    b = row['MktCap']
    c = row['previous_year']
    
    peer= find_matching_firm(mktcap,a,b,c)
    stock = {'IPO':ipo,'IPO return':2,'MFirm':peer,'MFirm return':4}
    peers=peers.append(stock, ignore_index=True)
    
peers
```
